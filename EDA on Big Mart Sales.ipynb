# Dealing with null values

```python
sales.isnull().sum()[sales.isnull().sum()!=0]
```

```python
sales.groupby('Item_Identifier')['Item_Weight'].median()
```

```python
sales['Item_Weight']=sales.groupby('Item_Identifier')['Item_Weight'].transform(lambda x:x.fillna(x.median()))
```

```python
sales.head()
```

```python
sales['Item_Weight']=sales.groupby('Item_Type')['Item_Weight'].transform(lambda x:x.fillna(x.median()))
```

There were around 2400 null values in Item weight we have sucessfully droped them. 

```python
filler=sales['Outlet_Size'].mode()[0]
filler
```

```python
sales['Outlet_Size'].fillna(filler,inplace=True)
```

```python
sales.isnull().sum()
```

We have successsfully droped 2410 null values from Outlet Size

# FEATURE ENGINEERING

```python
sales.head()
```

```python
sales['Item_Visibility']=sales.groupby('Item_Identifier')['Item_Visibility'].apply(lambda x:x.replace(0,x.median()))
```

```python
sales['Item_Identifier']=sales['Item_Identifier'].apply(lambda x:x[:2])
```

```python
sales.head()
```

```python
sales['Item_Fat_Content'].unique()
```

```python
sales['Item_Fat_Content'].replace({'LF':'Low Fat','low fat':'Low Fat','reg':'Regular'},inplace=True)
```

```python
sales.loc[sales.Item_Identifier=='NC','Item_Fat_Content']='Non Edible'
```

```python
sales['Item_Identifier'].unique()
```

```python
sales['Item_Type'].unique()
```

```python
perish=['Dairy','Meat','Fruits and Vegetables','Breakfast', 'Breads','Starchy Foods','Seafood']
```

```python
def perish_ables(x):
    if x in perish:
        return 'perishable'
    else:
        return 'non perishable'    
```

```python
sales['Item_Type']=sales.Item_Type.apply(perish_ables)
```

```python
sales.head()
```

```python
sales['Item_Type'].unique()
```

```python
sales.drop(columns=['Item_Type','Outlet_Establishment_Year'])
```

```python
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler()
```

```python
cols=['Item_Weight','Item_Visibility','Item_MRP','Item_Outlet_Sales']
```

```python
for k in cols:
    sales.loc[:,k]=sc.fit_transform(pd.DataFrame(sales.loc[:,k]))
```

```python
sales
```

```python
sales=pd.get_dummies(sales,drop_first=True)
```

```python
sales.drop('Outlet_Establishment_Year',axis=1,inplace=True)
```
